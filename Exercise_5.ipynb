```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 4 a)

```python
alpha = 0.001
x = -5
def df(x):
    return 4*x**3 + 15*x**2 - 7

path = np.array([x])
for k in range(200):
    x = x - alpha*df(x)
    path = np.append(path, x)
    
print(x)
```

```python
xx = np.linspace(-5, 2, 100)
yy = xx**4 + 5*xx**3 - 7*xx
plt.plot(xx, yy)
plt.plot(path, path**4 + 5*path**3 - 7*path, 'o')
plt.show()
```

# Exercise 4 b)

```python
alpha = 0.02
x = 15
y = 5
path_x = np.array([x])
path_y = np.array([y])

def dgx(x, y):
    return 6*x + 3*y + 3

def dgy(x, y):
    return 3*x + 4*y - 5

for k in range(1000):
    x1 = x - alpha*dgx(x, y)
    y1 = y - alpha*dgy(x, y)
    x = x1
    y = y1
    path_x = np.append(path_x, x)
    path_y = np.append(path_y, y)
    
print('x = ',x, 'y = ', y)
```

```python
xx = np.linspace(-15, 15, 100)
yy = np.linspace(-15, 15, 100)
(X, Y) = np.meshgrid(xx, yy)
Z = 3*X**2 + 3*X*Y + 2*Y**2 + 3*X - 5*Y
plt.contour(X, Y, Z)
plt.plot(path_x, path_y, 'r+')
plt.show()
```
